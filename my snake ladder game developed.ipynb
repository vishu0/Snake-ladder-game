```python
import random
    
class player():
    
    d={k: random.randint(1,100) for k in random.sample(range(0, 100), 10)}
    dd={(value if key<value else key):(key if key<value  else value) for key, value in d.items()  }
    l={k: random.randint(1,100) for k in random.sample(range(0, 100), 10)}
    ll= {(value if key>value else key):(key if key>value  else value) for key, value in l.items()  }
    c=0
    
    def __init__(self):
        self.name=input("Please Enter your name"+" ")
        self.old_val=0
        self.new_val=random.randint(1,6)
        self.win_point=100
        
        
    def snake(self):
        return player.dd
        
    def ladder(self):
        
        return player.ll
        
    
    def play(self):
        
       
        if player.c==0:
            
            print(self.name +" "+'GO:')
            val=self.old_val
            self.old_val=self.old_val+random.randint(1,6)
            input("pls press enter to role the dice")
            print("Rolling dice..." +  " " + "final value is"+" "+str(self.old_val))
            
            
            if self.old_val==self.win_point:
                val=self.old_val
                player.c=1
                print("CONGRATULATION " + " "+ self.name.upper()+ " " + "YOU ARE THE WINNER")
               
                
            
            if self.old_val!=self.win_point and self.c==0:
        
                if self.old_val>self.win_point:
                    print("Wait for next chance")
                    self.old_val=val
                    print('real value'+ str(val))
                
                if self.old_val==6:
                    self.old_val=self.old_val + random.randint(1,6)
                    input("wohoo...pls press enter again to role the dice you got one more chance")
                    print('roll the dice again..' + " value is " + " " + str(self.old_val))
                    
                    if (self.old_val)%6==0:
                        
                        self.old_val=self.old_val + random.randint(1,6)
                        print('roll the dice again..' + " value is " + " " + str(self.old_val))
                        #print(self.old)
                        
                if self.old_val in self.snake().keys():
                    self.old_val=self.snake().get(self.old_val)
                    print("snake bites~~~~~~>" + " "+ str(self.old_val))
                    
                    
                if self.old_val in self.ladder().keys():
                    #print("old val",self.old_val)
                    self.old_val=self.ladder().get(self.old_val)
                    print("woohooooo ladder"+ " "+ str(self.old_val))
                    
                
            
                
                
p1=player()
p2=player()
while (p1.old_val!=p1.win_point or p2.old_val!=p2.win_point) and player.c!=1:
    p1.play()
    p2.play()    
    
    
            
    


    
    
```
